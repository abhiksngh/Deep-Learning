```
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


#Deep Learning libraries
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```
data=pd.read_csv('Churn_Modelling.csv')

data.shape

X=data.iloc[:,3:13].values
y=data['Exited']
```

```
LabelEncoder_X_Geo=LabelEncoder()
X[:, 1]=LabelEncoder_X_Geo.fit_transform(X[:, 1])

LabelEncoder_X_Sex=LabelEncoder()
X[:, 2]=LabelEncoder_X_Sex.fit_transform(X[:, 2])

onehotencoder=OneHotEncoder(categorical_features=[1])
X=onehotencoder.fit_transform(X).toarray()

X=X[:,1:]
```

```

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=17)

X_train.shape

sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)

model=Sequential()
```

```



#Hidden Layers with activation as rectifier func
model.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
model.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
#Output Layer with sigmoid function
model.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

model.compile(optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy'])
```

```



model.fit(X_train,
    y_train,
    batch_size=10,
    epochs=400)
```

```
y_pred=model.predict(X_test)
```

```
y_pred=(y_pred>0.5)
```

```
from sklearn.metrics import confusion_matrix
```

```
confusion_matrix(y_test,y_pred)
```
